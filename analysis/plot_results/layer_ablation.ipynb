```python
import pandas as pd
import matplotlib.pyplot as plt

# 数据准备
data = {
    'Router': ['SMoE', 'RMoE', 'RMoE-NP', 'RMoE-NP-r0.5',
               'SMoE', 'RMoE', 'RMoE-NP', 'RMoE-NP-r0.5',
               'SMoE', 'RMoE', 'RMoE-NP', 'RMoE-NP-r0.5',
               'SMoE', 'RMoE', 'RMoE-NP', 'RMoE-NP-r0.5',
               'SMoE', 'RMoE', 'RMoE-NP', 'RMoE-NP-r0.5'],
    'layer': [6, 6, 6, 6,
              12, 12, 12, 12,
              18, 18, 18, 18,
              24, 24, 24, 24,
              32, 32, 32, 32],
    'val': [1.214, 1.213, 1.215, 1.217,
            1.158, 1.151, 1.169, 1.152,
            1.133, 1.125, 1.142, 1.133,
            1.128, 1.112, 1.149, 1.126,
            1.110, 1.100, 1.114, 1.101],
    'test': [1.184, 1.183, 1.185, 1.188,
             1.132, 1.125, 1.141, 1.127,
             1.109, 1.101, 1.107, 1.108,
             1.108, 1.089, 1.123, 1.101,
             1.085, 1.078, 1.090, 1.080]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(14, 6))

# plt.plot(all_smoe_gate_weights_norm, label='SMoE', linestyle='--')
# plt.plot(all_rmoe_gate_weights_norm, label='RMoE', linestyle='-.')
# plt.plot(all_rmoe_np_gate_weights_norm, label='RMoE_NP', linestyle=':')
# plt.plot(all_rmoe_np_r05_gate_weights_norm, label='RMoE_NP_r05', linestyle='-')

style_dict = {
    'SMoE': '--',
    'RMoE': '-',
    'RMoE-NP': ':',
    'RMoE-NP-r0.5': '-.'

}

# 绘制 val 值图
# plt.subplot(1, 2, 1)
plt.figure()
for router in df['Router'].unique():
    subset = df[df['Router'] == router]
    if router == 'RMoE':
        plt.plot(subset['layer'], subset['val'], label=router, linestyle=style_dict[router], marker='*')
    else:
        plt.plot(subset['layer'], subset['val'], label=router, linestyle=style_dict[router]) # marker='o',

plt.xlabel('Model Size (Layer)', fontsize=12)
plt.ylabel('Validation Results (BPC)', fontsize=12)
plt.title('Validation BPC vs Layer', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.savefig(f'layer_number_change_val.pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()

# 绘制 test 值图
# plt.subplot(1, 2, 2)
plt.figure()
for router in df['Router'].unique():
    subset = df[df['Router'] == router]
    if router == 'RMoE':
        plt.plot(subset['layer'], subset['val'], label=router, linestyle=style_dict[router], marker='*')
    else:
        plt.plot(subset['layer'], subset['val'], label=router, linestyle=style_dict[router]) # marker='o',

plt.xlabel('Model Size (Layer)', fontsize=12)
plt.ylabel('Test Results (BPC)', fontsize=12)
plt.title('Test BPC vs Layer', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.savefig(f'layer_number_change_test.pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
